```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fbprophet import Prophet
%matplotlib inline
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("https://fantasymovieleague.com/researchvault?section=box-office")
soup = BeautifulSoup(r.content, 'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))[0]
df.drop(df.columns[0], axis=1, inplace=True)
df.columns = df.columns.droplevel()
prices = df.iloc[:,0].str.split("$", n=0, expand=True)[1]
names = df.iloc[:,0].str.split("FB", n=0, expand=True)[0]
df["price"] = prices
df["name"] = names
firststep = df.drop(df.columns[0], axis = 1)
firststep.head()
```

```python
firststep["day"] = 'ALL'
cleaned = firststep
days = ['FRI','SAT','SUN']
cleaned.loc[cleaned.name.str.contains('|'.join(days)),"day"] = cleaned[cleaned.name.str.contains('|'.join(days))].name.str.split(" - ", n=0, expand=True)[0]
cleaned.loc[cleaned.name.str.contains('|'.join(days)),"name"] = cleaned[cleaned.name.str.contains('|'.join(days))].name.str.split(" - ", n=0, expand=True)[1]
cleaned.columns.values[[0,2,4,6]] = ["lastweek","2weeksago","3weeksago","4weeksago"]
final = cleaned.drop(cleaned.columns[[1,3,5,7]], axis = 1)
final
```

```python
final["url_part"] = final["name"]
the_mask = final["url_part"].str.startswith("The ")
final.loc[the_mask, "url_part"] = final.loc[the_mask, "url_part"].str.replace("The ","") + " The"
final["url_part"] = final["url_part"].str.replace(' ', '-')
final["url_part"] = final["url_part"].str.replace(': |,|\'|!','')
final["url_part"] = final["url_part"].str.replace(':', '-')
final
```

```python
def get_table(row):
    try:
        r = requests.get("https://www.the-numbers.com/movie/{}".format(row))
        soup = BeautifulSoup(r.content, 'lxml')
        table = soup.find_all(id = 'box_office_chart')[1]   
        movie = pd.read_html(str(table))[0]
        movie.columns = movie.iloc[0]
        movie = movie[1:]
        movie["Gross"] = movie["Gross"].str.replace('\$|,','')
        print("Found and stored data for {}!".format(row))
        return movie
    except IndexError:
        print("No luck finding {}, will try adding -(2018)".format(row))
        try:
            r = requests.get("https://www.the-numbers.com/movie/{}-(2018)".format(row))
            soup = BeautifulSoup(r.content, 'lxml')
            table = soup.find_all(id = 'box_office_chart')[1]   
            movie = pd.read_html(str(table))[0]
            movie.columns = movie.iloc[0]
            movie = movie[1:]
            movie["Gross"] = movie["Gross"].str.replace('\$|,','')
            print("Found and stored data for {}!".format(row))
            return movie
        except IndexError:
            print("No luck finding {}-(2018) either. Will try adding -(2017)".format(row))
            try:
                r = requests.get("https://www.the-numbers.com/movie/{}-(2017)".format(row))
                soup = BeautifulSoup(r.content, 'lxml')
                table = soup.find_all(id = 'box_office_chart')[1]   
                movie = pd.read_html(str(table))[0]
                movie.columns = movie.iloc[0]
                movie = movie[1:]
                movie["Gross"] = movie["Gross"].str.replace('\$|,','')
                print("Found and stored data for {}!".format(row))
                print(movie)
                return movie
            except IndexError:
                print("Looks like {} doesn't exist on the-numbers.com. Try a google search?".format(row))
```

```python
def forecast(df):
    df["y"] = np.log(pd.to_numeric(df["Gross"]))
    df["ds"] = df["Date"]
    m = Prophet()
    m.fit(df)
    future_data = m.make_future_dataframe(periods=7)
    forecast_data = m.predict(future_data)
    m.plot(forecast_data)
    plt.title(row)
    plt.show()
```
